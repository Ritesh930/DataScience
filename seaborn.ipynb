```python
## seaborn is a library for making statistical graphics in Python. it is built on top of matplotlib and 
# closely integrated with pandas data structures.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")  #tips is a pre loaded data set already there in seaborn with install
df.head()
```

```python
plt.plot(df['total_bill'])
plt.xlabel('Index')
plt.ylabel('Total Bill')
plt.title('Total Bill Plot')
plt.show()
```

```python
sns.lineplot(df['total_bill']) #plot graph
plt.title("bill graph")
plt.show()
```

```python
sns.scatterplot(df['total_bill'])
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
sns.histplot(df['total_bill'],color="green",alpha=0.5,edgecolor="black")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.kdeplot(df["total_bill"])
plt.show()
```

```python
#sns--1 categorical
# 2 numerical---scatter plot,line plot---relational---relplot
```

```python
sns.relplot(x="total_bill", y="tip", data=df,
            kind="scatter",
            hue="smoker",style="size",col="day",
            )
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=df,
            kind="swarm")
plt.show()
```

```python
sns.countplot(x="day", data=df)
plt.show()
```

```python
#pointplot--shows the mean of the data connected by lines toshow the trend.

sns.pointplot(x="day",y="total_bill",data=df,hue="sex")
```

```python
#boxplot--spread of data in the form of box.
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
#violinplot---spread of data in the form of violin
sns.violinplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.pairplot(df)  #charts on all numerical features.
plt.show()
```

```python
sns.histplot(df["tip"],kde=True)  #kde shows combination of graph.kde used to show density plot in histogram
plt.show()
```

```python
sns.boxplot(df["tip"])
plt.show()
```

```python
##outliers handle
q1=df["total_bill"].quantile(0.25)
q3=df["total_bill"].quantile(0.75)
iqr=q3-q1  #inter quartile range
lower_limit=q1-1.5*iqr #lower limit
upper_limit=q3+1.5*iqr #upper limit
print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
#df[df["tip"]<lower_limit]  #outliers below lower limit
df[df["total_bill"]>upper_limit]  #outliers above upper limit
#removing outliers
df=df[(df["total_bill"]>lower_limit) & (df["total_bill"]<upper_limit)]
sns.boxplot(df["total_bill"])
plt.show()
```
